# Feature Engineering 
ADD columns

```python
data["Attack"].mean()
```

```python
attack_mean = data["Attack"].mean()

def set_attack(val):
    if val < attack_mean:
        return "Attack Low"
    elif val == attack_mean:
        return "Attack neutral"
    else:
        return "Attack High"
     
```

```python
# "Attack_high_low"  -> "Attack High" , "Attack Low"
data["Attack_high_low"] = data["Attack"].apply(set_attack)
data
```

# Assignment 

```python
speed_mean = data["Speed"].mean()

def set_Speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed neutral"
    else:
        return "Speed high"
```

```python
# "Speed_high_low"  -> "Speed High" , "Speed Low"
data["Speed_high_low"] = data["Speed"].apply(set_Speed)
data
```

```python
hp_mean = data["HP"].mean()

def set_hP(val):
    if val < hp_mean:
        return "HP Low"
    elif val == hp_mean:
        return "HP neutral"
    else:
        return "HP High"
```

```python
# "HP_high_low"  -> "HP High" , "HP Low"
data["HP_high_low"] = data["HP"].apply(set_HP)
data
```
